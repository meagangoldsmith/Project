```python
##EXTRACTING FROM SQL DATABASE
##using files created on sql to make dataframe with pandas
characters = pd.read_csv('dunech.csv')
movies = pd.read_csv('dunemovies.csv')
salaries= pd.read_csv('salariesdune.csv')
```

```python
###TRANSFORMING
###Replacing the 0 dollars with actors and characters to say n.a
salaries['salary'] = salaries['salary'].replace(0, 'n.a.')
salaries.head()
```

```python
###TRANSFORMING
##Replacing language throughout the entire dataframe
characters = characters.replace('Non','Not in Book')
characters = characters.replace('Yes','In Book')
characters.head()
```

```python
###TRANSFORMING
###Replacing language throughout the entire dataframe for in movie or not
movies = movies.replace('Non','Not in Movie')
movies = movies.replace('Yes','In Movie')
movies.head()
```

```python
###TRANSFORMING
###Replacing language throuhgout the entire dataframe
reviews['Review Date'] = pd.to_datetime(reviews['Review Date'])
reviews['Review Date'] = reviews['Review Date'].dt.strftime('%m.%d.%Y')

reviews.head()
```

```python
###LOADING
import sqlite3
###Import to make a cloud database
conn = sqlite3.connect('dune.db')
cursor = conn.cursor()
###Using the transformed dataframes to create the new Dune database
reviews.to_sql('reviews', conn, if_exists='replace', index=False)
movies.to_sql('movies', conn, if_exists='replace', index=False)
characters.to_sql('characters', conn, if_exists='replace', index=False)
salaries.to_sql('salaries', conn, if_exists='replace', index=False)
conn.commit()
conn.close()
```
